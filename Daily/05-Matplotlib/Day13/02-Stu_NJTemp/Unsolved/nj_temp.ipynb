```python
#import libraries
from typing import List, Any, Union
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Average temperature per month in Fahrenheit: `[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
avgTemp=[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
avgTemp
```

```python
#Assign to the x-axis a range of numerical values representing each month of the year.
x_axis=np.arange(1,13,1)
x_axis
```

```python
#Plot the Fahrenheit points.
plt.plot(x_axis,avgTemp)
plt.show()
```

```python
#Use a list comprehension to convert the temperature to Celsius.
Cel_avgTemp=[(x-32)*0.56 for x in avgTemp]
Cel_avgTemp
```

```python
#* Plot the Celsius points.
plt.plot(x_axis,Cel_avgTemp)
plt.show()
```

```python
#create a third plot with both the Fahrenheit and Celsius points.
plt.plot(x_axis,avgTemp,label='Fahrenheit',color='green')
plt.plot(x_axis,Cel_avgTemp,label='Celcius', color='orange')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.title('Monthly Averages for Temperature in New Jersey')
plt.legend(loc='best')
plt.show()

```
