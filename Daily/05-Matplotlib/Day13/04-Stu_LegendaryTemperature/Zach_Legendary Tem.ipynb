```python
from typing import List, Any, Union
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline
```

```python
#Average temperature per month in Fahrenheit: `[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
avgTemp=[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
avgTemp
```

```python
CTemp=[(x-32)*.56 for x in avgTemp]
CTemp
```

```python
#x-axis
xaxis=np.arange(1,13,1)
xaxis
```

```python
plt.plot(xaxis,avgTemp)
plt.show()
```

```python

plt.plot(xaxis,avgTemp,color='blue',label='Fahrenhei',marker='o')
plt.plot(xaxis,CTemp,color='red',label='Celcius',marker='x')
plt.xlabel('Months')
plt.ylabel('Degrees')
plt.title('Monthly Averages for Temperature in New Jersey')
plt.legend(loc='best')
```
