```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger=[9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail=[75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
danger
```

```python
x_axis=np.arange(0,130,10)
x_axis
```
