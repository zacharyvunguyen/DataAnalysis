```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load = pd.read_csv("data/ride_data.csv")
```

```python
# Read the City Data
city_data_to_load.head()
city_data_to_load.shape
city_data_to_load.columns
```

```python
# Read the Ride Data
ride_data_to_load.head()
ride_data_to_load.shape
ride_data_to_load.columns
```

```python
# Combine the data into a single dataset
combine_data=pd.merge(ride_data_to_load,city_data_to_load,how='left',on='city')
combine_data.shape
```

```python
# Display the data table for preview
combine_data.head(2)
```

```python
#AVERAGE FARE PER CITY
avg_fare_city=combine_data.groupby('city').fare.mean()
avg_fare_city.head()
```

```python
#TOTAL NUMBER OF RIDE PER CIRY
total_ride_city=combine_data.groupby('city').ride_id.count()
total_ride_city.head()
```

```python
#TOTAL NUMBER OF DRIVER PER CITY
total_driver_city=combine_data.groupby('city').driver_count.first()
total_driver_city.head()
```

```python
##CITY TYPE
city_type=combine_data.groupby('city').type.first()
```

```python
combine_data['type_number']=combine_data['type'].map({"Rural":0,"Suburban":1,"Urban":2})
colors=combine_data.groupby('city').type_number.first()
colors.head()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(total_ride_city,avg_fare_city,alpha=0.5,s=total_driver_city*30,c=colors,edgecolors='black')
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City")
plt.ylabel("Average Fare ($)")
plt.legend()
plt.show()
```

```python
######TOTAL FARES BY CITY TYPE
total_fares_type=combine_data.groupby('type').fare.sum()
total_fares=total_fares_type.sum()
percent_fares_type=(total_fares_type)*100/total_fares
percent_fares_type
```

```python
####BACKUP DATA
df1 = pd.DataFrame()
df1['average_fare'] = avg_fare_city
df1['ride_number'] =total_ride_city
df1['driver_number']=total_driver_city
df1['city_type']=city_type
df1['type_number']=df1['city_type'].map({"Rural":0,"Suburban":1,"Urban":2})
df1.head()

## Bubble Plot of Ride Sharing Data
```

```python
plt.pie(df1.driver_number.mean(),labels=df1['driver_number'].groupby('city_type'))
plt.show()

```
