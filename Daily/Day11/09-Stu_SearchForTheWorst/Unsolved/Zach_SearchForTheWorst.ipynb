```python
import pandas as pd
```

```python
path='Resources/Soccer2018Data.csv'
```

```python
#read csv
df=pd.read_csv(path,low_memory=False)
df.head()
```

```python
# Collect a list of all the unique values in "Preferred Position"
df["Preferred Position"].unique()
```

```python
# Looking only at strikers (ST) to start
df_st=df.loc[df['Preferred Position']=="ST"]
df_st.head()
```

```python
#Sort the DataFrame by the values in the "ST" column to find the worst
worst=df_st.sort_values(by='ST',ascending=True)
worst.head()
```

```python
# Reset the index so that the index is now based on the sorting locations
worst_reset=worst.reset_index(drop=True)
worst_reset
```

```python
# Save all of the information collected on the worst striker
worst_guy=worst_reset.loc[0,:]
worst_guy
```
